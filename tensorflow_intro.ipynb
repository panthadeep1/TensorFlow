```python
import tensorflow as tf

# Build a dataflow graph.
c = tf.constant([[1.0, 2.0], [3.0, 4.0]])
d = tf.constant([[1.0, 1.0], [0.0, 1.0]])
print(f"c: {c}")
print(f"d: {d}")

e = tf.matmul(c, d)
print(f"e: {e}")
```
